```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
hundred_enc_out = pd.read_excel('enc_out.xlsx').iloc[:,1::]
ninety_enc_out = pd.read_excel('ninety_enc_out.xlsx').iloc[:,1::]
eighty_enc_out = pd.read_excel('eighty_enc_out.xlsx').iloc[:,1::]
seventy_enc_out = pd.read_excel('seventy_enc_out.xlsx').iloc[:,1::]
sixty_enc_out = pd.read_excel('sixty_enc_out.xlsx').iloc[:,1::]
gen_enc_out = pd.read_excel('gen_enc_out.xlsx').iloc[:,1::]
dqn_enc_out = pd.read_excel('dqn_enc_out.xlsx').iloc[:,1::]
```

```python
dqn_enc_out[0][0]
```

```python
from matplotlib import animation

fig = plt.figure(figsize=(20,20))
ax = fig.add_subplot(111, projection='3d')

def init():
    ax.scatter(hundred_enc_out[0], hundred_enc_out[1], hundred_enc_out[2],
                    edgecolor='none', s=15, alpha=0.3)

    ax.scatter(ninety_enc_out[0], ninety_enc_out[1], ninety_enc_out[2],
                    edgecolor='none', s=15, alpha=0.3)

    ax.scatter(eighty_enc_out[0], eighty_enc_out[1], eighty_enc_out[2],
                    edgecolor='none', s=15, alpha=0.3)

    ax.scatter(seventy_enc_out[0], seventy_enc_out[1], seventy_enc_out[2],
                    edgecolor='none', s=15, alpha=0.3)

    ax.scatter(sixty_enc_out[0], sixty_enc_out[1], sixty_enc_out[2],
                    edgecolor='none', s=15, alpha=0.3)

    ax.scatter(gen_enc_out[0], gen_enc_out[1], gen_enc_out[2],
                    edgecolor='none', s=15, alpha=0.3)
    
    ax.scatter(dqn_enc_out[0], dqn_enc_out[1], dqn_enc_out[2],
                    edgecolor='none', s=15, alpha=0.3)
    return fig, 

def animate(i):
    ax.view_init(elev=30., azim=i)
    return fig,

# Animate
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=360, interval=20, blit=True)
```

```python
anim.save('temp.gif', writer='pillow', fps=30)
```
