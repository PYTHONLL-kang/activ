```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler, Normalizer
```

```python
df = pd.read_excel('aug_nine_var.xlsx')
```

```python
df = df.iloc[:, 1::]
```

```python
scaler = StandardScaler()
sta_df = scaler.fit_transform(df)

scaler = MinMaxScaler()
mm_df = scaler.fit_transform(df)

scaler = RobustScaler()
rob_df = scaler.fit_transform(df)

scaler = Normalizer()
nom_df = scaler.fit_transform(df)
```

```python
sta_df = pd.DataFrame(sta_df)
mm_df = pd.DataFrame(mm_df)
rob_df = pd.DataFrame(rob_df)
nom_df = pd.DataFrame(nom_df)

data_list = []
sta_data_list =[]
mm_data_list =[]
rob_data_list =[]
nom_data_list =[]
```

```python
for i in range(len(df.columns)-1):
    data_list.append(df.iloc[:,i:i+1])

for i in range(len(df.columns)-1):
    sta_data_list.append(sta_df.iloc[:,i:i+1])

for i in range(len(df.columns)-1):
    mm_data_list.append(mm_df.iloc[:,i:i+1])

for i in range(len(df.columns)-1):
    rob_data_list.append(rob_df.iloc[:,i:i+1])

for i in range(len(df.columns)-1):
    nom_data_list.append(nom_df.iloc[:,i:i+1])
```

```python
print(data_list[0])
```

```python
list = data_list

f, axes = plt.subplots(3, 7)

f.set_size_inches((10, 7))

plt.subplots_adjust(wspace = 0.3, hspace = 0.3)
plt.tick_params(left = False, right = False , labelleft = False ,
                labelbottom = False, bottom = False)

x = 0
for i in range(3):
    for j in range(7):
        axes[i][j].plot(list[x])
        x = x + 1

plt.savefig('re_data_structure.png')
    
plt.show()
```

```python
list = [data_list, sta_data_list, mm_data_list, rob_data_list, nom_data_list]

f, axes = plt.subplots(21, 5)

f.set_size_inches((10, 7))

plt.subplots_adjust(wspace = 0.3, hspace = 0.3)
plt.tick_params(left = False, right = False , labelleft = False ,
                labelbottom = False, bottom = False)

for i in range(21):
    for j in range(5):
        axes[i][j].plot(list[j][i])

plt.savefig('re_data_structure.png')
    
plt.show()
```
