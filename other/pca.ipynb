```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
np.set_printoptions(precision=6, suppress=True)
```

```python
def learn_pca(df):
    pca = PCA(n_components=2)
    pca.fit(df)
    df_pca = pca.transform(df)
    df_pca = pd.DataFrame(df_pca, columns = [str(i) for i in range(0,2)])
    # df_stock = df_stock.astype(float)
    return df_pca
```

```python
a = pd.read_excel("real_Result/basic.xlsx")
b = pd.read_excel("real_Result/basic_formula.xlsx")
c = pd.read_excel("real_Result/reinforce.xlsx")
d = pd.read_excel("real_Result/result.xlsx")
e = pd.read_excel("real_Result/result_not_consider.xlsx")
f = pd.read_excel("real_Result/autoencoder_result.xlsx")
g = pd.read_excel("real_Result/dqn_result.xlsx")
real = pd.read_excel('real_Result/aug_nine_var.xlsx')

a = pd.DataFrame(a.iloc[:,1:])
b = pd.DataFrame(b.iloc[:,1:])
c = pd.DataFrame(c.iloc[:,1:])
d = pd.DataFrame(d.iloc[:,1:])
e = pd.DataFrame(e.iloc[:,1:])
f = pd.DataFrame(f.iloc[:,1:])
g = pd.DataFrame(g.iloc[:,1:])
real = pd.DataFrame(real.iloc[:,1:])
```

```python
a_ = learn_pca(a)
b_ = learn_pca(b)
c_ = learn_pca(c)
d_ = learn_pca(d)
e_ = learn_pca(e)
f_ = learn_pca(f)
g_ = learn_pca(g)
real_ = learn_pca(real)
```

```python
plt.figure(figsize=(15,10))
plt.scatter(a_.iloc[:,0:1], a_.iloc[:,1::], c='b', s=1)
plt.scatter(b_.iloc[:,0:1], b_.iloc[:,1::], c='r', s=1)
plt.scatter(c_.iloc[:,0:1], c_.iloc[:,1::], c='orange', s=1)
plt.scatter(d_.iloc[:,0:1], d_.iloc[:,1::], c='c', s=1)
plt.scatter(e_.iloc[:,0:1], e_.iloc[:,1::], c='g', s=1)
plt.scatter(f_.iloc[:,0:1], f_.iloc[:,1::], c='m', s=1)
plt.scatter(g_.iloc[:,0:1], g_.iloc[:,1::], c='k', s=1)
plt.scatter(real_.iloc[:,0:1], real_.iloc[:,1::], c='y', s=1)

plt.legend(['basic', 'basic_formula', 'reinforce', 'result', 'result_not_consider', 'ae_result', 'dqn_result'])
plt.savefig('vis.png')
plt.show()
```
