```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
import os
os.chdir("C:\\code\\activ")
```

```python
df_name = 'nov_nine_var.xlsx'
```

```python
np.set_printoptions(precision=6, suppress=True)
```

```python
def learn_pca(df):
    pca = PCA(n_components=2)
    pca.fit(df)
    df_pca = pca.transform(df)
    df_pca = pd.DataFrame(df_pca, columns = [str(i) for i in range(0,2)])
    # df_stock = df_stock.astype(float)
    return df_pca
```

```python
path = "./documents/result/"
a = pd.read_excel(path+"basic.xlsx")
b = pd.read_excel(path+"basic_formula.xlsx")
c = pd.read_excel(path+"reinforce.xlsx")
f = pd.read_excel(path+"ae_formula.xlsx")
g = pd.read_excel(path+"dqn.xlsx")
real = pd.read_excel('./documents/'+df_name)

a = pd.DataFrame(a.iloc[:,1:])
b = pd.DataFrame(b.iloc[:,1:])
c = pd.DataFrame(c.iloc[:,1:])
f = pd.DataFrame(f.iloc[:,1:])
g = pd.DataFrame(g.iloc[:,1:])
real = pd.DataFrame(real.iloc[:,1:])
```

```python
a_ = learn_pca(a)
b_ = learn_pca(b)
c_ = learn_pca(c)
f_ = learn_pca(f)
g_ = learn_pca(g)
real_ = learn_pca(real)
```

```python
plt.figure(figsize=(15,10))
plt.scatter(a_.iloc[:,0:1], a_.iloc[:,1::], c='b', s=1)
plt.scatter(b_.iloc[:,0:1], b_.iloc[:,1::], c='r', s=1)
plt.scatter(c_.iloc[:,0:1], c_.iloc[:,1::], c='orange', s=1)
plt.scatter(f_.iloc[:,0:1], f_.iloc[:,1::], c='m', s=1)
plt.scatter(g_.iloc[:,0:1], g_.iloc[:,1::], c='k', s=1)
plt.scatter(real_.iloc[:,0:1], real_.iloc[:,1::], c='y', s=1)

plt.legend(['basic', 'basic_formula', 'reinforce', 'result', 'result_not_consider', 'ae_result', 'dqn_result'])
plt.savefig('vis.png')
plt.show()
```
